```
import pandas as pd
```

```
# Crear diccionario con datos
datos = {
    "Producto": ["Camara ip", "Camara wifi", "Camara analogica", "Domo PTZ", "Camara 4 lentes"],
    "Precio": [250000, 155000, 95000, 1100000, 890000],
    "Cantidad": [32, 15, 25, 28, 17]
}

# Crear DataFrame
df = pd.DataFrame(datos)

# Mostrar DataFrame
print(df)
```

```
# Operación entre columnas
df["Total"] = df["Precio"] * df["Cantidad"]

# Mostrar resultado
print(df)
```

```
# Aplicar descuento del 16%
df["Descuento"] = df["Total"] * 0.16

# Total final
df["Total_Final"] = df["Total"] - df["Descuento"]

print(df)
```
