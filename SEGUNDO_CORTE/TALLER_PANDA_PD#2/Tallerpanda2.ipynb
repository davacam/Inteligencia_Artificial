# **2. OPERACIONES MATEMÁTICAS ENTRE COLUMNAS**

```python
df["Total_Venta"] = df["Precio"] * df["Cantidad"]
df.head()
```

```python
df["Costo_Total"] = df["Total_Venta"] + df["Costo_Envio"]
df.head()
```

```python
#CALCULO DEL PROMEDIO TOTAL DE VENTAS CON LA MEDIDA CENTRAL MEAN-PROMEDIO
promedio_ventas = df["Total_Venta"].mean()

promedio_ventas
```

```python
mayor_venta = df.loc[df["Total_Venta"].idxmax()]

print("Producto con mayor Total_Venta")
print(mayor_venta)
```

```python
menor_venta = df.loc[df["Total_Venta"].idxmin()]

print("Producto con menor Total_Venta")
print(menor_venta)
```

```python
df["IVA"] = df["Total_Venta"] * 0.19

print(df.head())
```

```python
df["Ganancia_Estimada"] = df["Total_Venta"] * 0.25

print(df.head())
```

```python
df["Perdida_Estimada"] = df["Total_Venta"] * 0.0005

print(df.head())
```

```python
total_general = df["Total_Venta"].sum()

df["Porcentaje_Ventas"] = (df["Total_Venta"] / total_general) * 100

print(df.head())
```

```python
productos_15 = df[df["Porcentaje_Ventas"] > 15]

print("Productos con porcentaje superior al 15%")
print(productos_15)
```

```python
media_precio = df["Precio"].mean()
mediana_precio = df["Precio"].median()
moda_precio = df["Precio"].mode()

print("Media:", media_precio)
print("Mediana:", mediana_precio)
print("Moda:")
print(moda_precio)
```

```python
desviacion = df["Cantidad"].std()

print("Desviación estándar:")
print(desviacion)
```

```python
maximo = df["Total_Venta"].max()
minimo = df["Total_Venta"].min()

print("Valor máximo:", maximo)
print("Valor mínimo:", minimo)
```
